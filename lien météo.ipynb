```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt 
import requests
import seaborn as sns
sns.set(style="white") #white background style for seaborn plots
sns.set(style="whitegrid", color_codes=True)
```

```python
u = 'https://donneespubliques.meteofrance.fr/metadonnees_publiques/fiches/fiches.json'
r = requests.get(u)
print(r.status_code)
r.json()
```

```python
d = r.json()
d
```

```python
for elem in d['features']:
    print('#################################')
    print(elem["properties"]['COMMUNE'],elem["properties"]['ficheClimReduite'],elem["properties"]['NUM_POSTE'])
    print("https://donneespubliques.meteofrance.fr//FichesClim/FICHECLIM_"+elem["properties"]['NUM_POSTE']+".data")
    
    url="https://donneespubliques.meteofrance.fr/FichesClim/FICHECLIM_"+elem["properties"]['NUM_POSTE']+".data"
    print(url)

    if elem["properties"]['ficheClimReduite']==1:
        r2 = requests.get(url, allow_redirects=True)
        open(elem["properties"]['COMMUNE']+".csv", 'wb').write(r2.content)

    
```
