```python
Pop = pd.read_csv("/content/TM.csv")


Pop['Date'] = pd.to_datetime(Pop['Date'], infer_datetime_format=True)
Pop = Pop.sort_values(by='Date')

Pop = Pop.set_index(['Date'])

Prix = Pop['prix moyen au kg']

Production = Pop['Production quantit√© \ntonne(s)']
```

```python
p =[0, 1, 2, 3, 4, 5, 6]
d = range(0, 3)
q = [0, 1, 2, 3]
```

**Prix**

```python
def p_prix(Prix, p, d, q):
  data = pd.DataFrame()
  for i in p:
    for j in d:
      for k in q:
        order = (i,j,k)
      try:
        mod = ARIMA(Prix,order)
        results = mod.fit()
        # print(results.summary())
        RMSEP = sqrt(mean_squared_error(Prix, results.predict()))
        print(RMSEP)
        MaeP = mean_absolute_error(Prix, results.predict())
        print(MaeP)
        data = data.append({"p":i, "d":j,"q":k,"RMSE":RMSEP , "MAE":MaeP}, ignore_index=True)
       
        
      except:
        pass
  data = data.sort_values(by='RMSE')
  return data  
```

```python
p_prix(Prix, p, d, q)
```
