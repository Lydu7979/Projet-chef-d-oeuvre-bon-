```python
#Importation des librairies nécessaires

import streamlit as st
import pandas as pd
import numpy as np
import chart_studio.plotly as plotly
import plotly.figure_factory as ff
from plotly import graph_objs as go
from fbprophet import Prophet
from fbprophet.plot import plot_plotly
```

```python
st.title('App Prix Production de tomates')

dataset = ('TM','TM2')

option = st.selectbox('Choisir le dataset pour les prédictions',dataset)
DATA_URL =('./DATA/'+option+'.csv')

Année = st.slider('Année pour les prédicitions:',1,8)
Jour = st.slider('Jour pour les prédictions:',1,30)

@st.cache
def load_data():
    data = pd.read_csv(DATA_URL)
    return data



data_load_state = st.text('Téléchargement des données...')
data = load_data()
data_load_state.text('Téléchargement des données... terminé!')

def plot_fig():
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=data.Date, y=data['prix moyen au kg'], name="prix moyen au kg",line_color='darkviolet'))
    fig.add_trace(go.Scatter(x=data.Date, y=data['Production quantité \ntonne(s)'], name="Production quantité \ntonne(s)",line_color='gold'))
    fig.layout.update(title_text='Time Series data with Rangeslider',xaxis_rangeslider_visible=True)
    st.plotly_chart(fig)
    return fig

if st.checkbox('Show raw data'):
    st.subheader('Raw data')
    
plot_fig()


data_pred1 = data[['Date','prix moyen au kg']]
data_pred1['ds'] = pd.to_datetime(data_pred1['Date'])
data_pred1['y'] = data_pred1['prix moyen au kg']
data_pred1 = data_pred1.drop(["Date","prix moyen au kg"], axis=1)

data_pred2 = data[['Date','Production quantité \ntonne(s)']]
data_pred2['ds'] = pd.to_datetime(data_pred2['Date'])
data_pred2['y'] = data_pred2['Production quantité \ntonne(s)']
data_pred2 = data_pred2.drop(["Date","Production quantité \ntonne(s)"], axis=1)

m = Prophet(daily_seasonality=True)
m.fit(data_pred1)
future = m.make_future_dataframe(periods=7, freq='D')
forecast = m.predict(future)


m2 = Prophet(daily_seasonality=True)
m2.fit(data_pred2)
future2 = m2.make_future_dataframe(periods=7, freq='D')
forecast2 = m.predict(future2)



fig1 = plot_plotly(m, forecast)
if st.checkbox('Show forecast data'):
    st.subheader('forecast data')
    st.write(forecast)
st.write('Forecasting closing of stock value for'+option+' pour une période de: '+str(Jour)+'Jour')
st.plotly_chart(fig1)

fig2 = plot_plotly(m2, forecast2)
if st.checkbox('Show forecast data'):
    st.subheader('forecast data')
    st.write(forecast2)
st.write('Forecasting closing of stock value for'+option+' pour une période de: '+str(Jour)+'Jour')
st.plotly_chart(fig2)


st.write("Component wise forecast")
fig3 = m.plot_components(forecast)
st.write(fig3)

st.write("Component wise forecast")
fig4 = m2.plot_components(forecast2)
st.write(fig4)
```
