```python
import numpy as np 
import pandas as pd 

from sklearn import preprocessing
from sklearn.model_selection import train_test_split


import matplotlib.pyplot as plt 

import seaborn as sns
sns.set(style="white") #white background style for seaborn plots
sns.set(style="whitegrid", color_codes=True)
```

```python
T = pd.read_csv("prix tomates allongées  coeur au kg  - Sheet1.csv")
T
```

```python
T.describe()
```

```python
Corr = T.corr()
print(Corr)

f, ax = plt.subplots(figsize=(5,5))
plt.title('Corrélation entre variables')


sns.heatmap(Corr,linewidths=0.25,vmax=1.0, square=True, cmap="summer", linecolor='black', annot=True)
plt.show()
```

```python
T['Date'].value_counts().to_frame()
```

```python
T['Marché ou le produit est vendu'].value_counts().to_frame()
```

```python
T['Marché ou le produit est vendu'].value_counts().plot(kind='bar',figsize=(30,5),title= 'marché',color ='darkviolet')
plt.xlabel('marché ')
plt.ylabel("nombre de fois ou le marché apparaît")
plt.show()
```

```python
T['Région'].value_counts().plot(kind='bar',figsize=(5,5),title= 'Région',color ='gold')
plt.xlabel('Région')
plt.ylabel("nombre de fois ou la région apparaît")
plt.show()
```

```python
T['Région'].value_counts().plot(kind='pie', title='Région', figsize=[5,5],
          autopct=lambda p: '{:.0f}%'.format(p,(p/100)*T['Région'].value_counts().sum()))
```

```python
T['Marché ou le produit est vendu'].value_counts().plot(kind='pie', title='Marché ou le produit est vendu', figsize=[10,10],
          autopct=lambda p: '{:.0f}%'.format(p,(p/100)*T['Marché ou le produit est vendu'].value_counts().sum()))
```
